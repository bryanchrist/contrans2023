```python
import numpy as np
import pandas as pd
import os
from bs4 import BeautifulSoup
import json
import xmltodict
import pymongo
from bson.json_util import loads, dumps
mongo_username = os.getenv('MONGO_INITDB_ROOT_USERNAME')
mongo_password = os.getenv('MONGO_INITDB_ROOT_PASSWORD')
mongo_init_db = os.getenv('MONGO_INITDB_DATABASE')
```

```python
myclient = pymongo.MongoClient(
    f"mongodb://{mongo_username}:{mongo_password}@mongo:27017/{mongo_init_db}?authSource=admin"
)
```

```python
contrans_db = myclient['contrans'] #access db
```

```python
bills = contrans_db['bills'] #access bill collection in db
```

```python
bills.count_documents({})
```

```python
myquery = bills.find({'sponsors.item.lastName': 'Good'}, {'title':1}) #dict 1 filters rows and dict 2 filters columns
bill_text= dumps(myquery)
bill_records = loads(bill_text)
bill_df = pd.json_normalize(bill_records)
[print(x) for x in bill_df['title']]
```

```python
myquery = bills.find({'sponsors.item.lastName': 'Good'}, {'title':1, 'type': 1, 'summaries.summary':1}) #dict 1 filters rows and dict 2 filters columns
bill_text= dumps(myquery)
bill_records = loads(bill_text)
bill_df = pd.json_normalize(bill_records)
bill_df
```

```python
bills.create_index([('title', 'text')]) #tells search engine we are looking in title column and looking for text
```

```python
myquery = bills.find({'$text': {'$search': 'artificial intelligence', "$caseSensitive": False}}) #dict 1 filters rows and dict 2 filters columns
bill_text= dumps(myquery)
bill_records = loads(bill_text)
bill_df = pd.json_normalize(bill_records)
bill_df['title']
```
